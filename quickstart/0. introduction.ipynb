## **SCAPES: Semantically Conditioned Auto-Regressive Prior for Environmental Sounds**

### 1. Introduction

SCAPES is a generative model developed at the Music Technology Group of Universitat Pompeu Fabra, designed to generate audio controlled by semantic embeddings. SCAPES is a fairly complex model built on top of several state-of-the-art technologies. In this course, we will not cover all the details, but it is important to understand at a high level what the model is and what it does.

#### 1.1 What is SCAPES?

SCAPES is a generative model based on a Continuous Normalizing Flow [1], a state-of-the-art type of generative model. The main architecture used is a self-attending Transformer [2], and it is trained using Flow Matching [3]. The model uses two conditioning mechanisms: cross-attention with past audio, and AdaLN with semantic embeddings computed via CLAP [4]. SCAPES does not generate audio directly; instead, it generates EnCodec embeddings [5], and is designed to generate small chunks of audio. To generate long audio segments, it must be used autoregressively.

[1] Chen, Ricky T. Q., Rubanova, Yulia, Bettencourt, Jesse, & Duvenaud, David. "Neural Ordinary Differential Equations." Advances in Neural Information Processing Systems (NeurIPS), 2018.

[2] Vaswani, Ashish, et al. "Attention is all you need." Proceedings of the 31st International Conference on Neural Information Processing Systems, 2017.

[3] Lipman, Yaron, et al. "Flow Matching for Generative Modeling." International Conference on Learning Representations, 2023.

[4] Wu, Yusong, et al. "Large-scale Contrastive Language-Audio Pretraining with Feature Fusion and Keyword-to-Caption Augmentation." IEEE International Conference on Acoustics, Speech and Signal Processing, ICASSP, 2023.

[5] Défossez, Alexandre, et al. "High Fidelity Neural Audio Compression." arXiv:2210.13438, 2022.

#### 1.2 What does SCAPES do?

In practice, SCAPES generates audio based on past audio and a condition that contains semantic (i.e., coarse) information about the audio to be generated. This condition can be controlled 10 times per second, giving the user fine-grained control over the generation process.
